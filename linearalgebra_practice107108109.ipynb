# Типовой расчет 7 / Standard Calculation 7

## Постановка  задачи / Formulation of the problem:

Выполнить указанные действия с комплексными числами, представив ответ в алгебраической форме:

$\large{Z_1 + Z_2, Z_2 - Z_3, Z_1 * Z_4, Z_2 / Z_5}$

$\large{Z_1 = 4 + 3 i; Z_2 = 1 + 2 i; Z_3 = 1 + 3 i; Z_4 = – 10 + i; Z_5 = – 7 + 6 i}$

```python
z1 = complex(4 + 3 * I)
z2 = complex(1 + 2 * I)
z3 = complex(1 + 3 * I)
z4 = complex(-10 + 1 * I)
z5 = complex(-7 + 6 * I)

z1, z2, z3, z4, z5
```

## Решение / Decision

Сложение комплексных чисел.

$(a + bi) + (c + di) = (a + c) + (b + d)i$

$Z_1 + Z_2 =  (4 + 3 i) + (1 + 2 i) = 5 + 5i$

Вычитание комплексных чисел.

$(a + bi) - (c + di) = (a - c) + (b - d)i$

$Z_2 - Z_3 =  (1 + 2 i) - (1 + 3i)= - i$

Умножение комплексных чисел.

$(a + bi) * (c + di) = (ac - bd) + (ad + bc)i$

$Z_1 + Z_4 =  (4 + 3 i) + (-10 + i) = (4 * (-10) - 3 * 1) + (4 * 1 + 3 * (-10))i = -43 -26i$

Деление комплексных чисел.

$\large{\frac{a + bi}{c + di} = \frac{(a + bi) * (c - di)}{(c + di) * (c - di)} = \frac{ac+ bd}{c^2 + d^2} + \frac{bc - ad}{c^2 + d^2}i}$

$\large{Z_2 / Z_5 = \frac{1 + 2i}{-7 + 6i} = \frac{1 * (-7) + 2 * 6}{(-7)^2 + 6^2} + \frac{2 * (-7) - 1 * 6}{(-7)^2 + 6^2}i} = \frac{1}{17} - \frac{4}{17}i$

# Типовой расчет  8 / Standard Calculation 8

## Постановка  задачи / Formulation of the problem:

По заданным комплексным числам $Z_1, Z_2, Z_3$ и показателям $K, L, M, N$ вычислить

$\large{Z_1^K * Z_2^L, \ Z_2^M \ / \ Z_3^N}$.

Ответ представить в алгебраической форме.

$Z_1 = -\sqrt{2 - \sqrt{3}} + \sqrt{2 + \sqrt{3}} i; \ Z_2 = \sqrt{3} + i; \ Z_3 = -1 + i$

$K = 10; \ L = 6; \ M = 15; \ N = 27$ 

## Решение / Decision

### I.

Для получения тригонометрической формы заданных комплексных чисел найдем их модули и аргументы.

---

$Z_1 = -\sqrt{2 - \sqrt{3}} + \sqrt{2 + \sqrt{3}} i$

$r_1 = |Z_1| = \sqrt{\Big(-\sqrt{2 - \sqrt{3}}\Big)^2 + \Big(\sqrt{2 + \sqrt{3}}\Big)^2} = \sqrt{2 - \sqrt{3} + 2 + \sqrt{3}} = \sqrt{4} = 2$

$-\sqrt{2 - \sqrt{3}} < 0; \sqrt{2 + \sqrt{3}} > 0 \implies \varphi_1 \in II $ четверти.

$\large{\varphi_1 = \pi - arctg\frac{\sqrt{2 + \sqrt{3}}}{\sqrt{2 - \sqrt{3}}} = \pi - arctg\frac{\sqrt{2 + \sqrt{3}} * \sqrt{2 + \sqrt{3}}}{\sqrt{2 - \sqrt{3}} * \sqrt{2 + \sqrt{3}}} = }$

$\large{= \pi - arctg \frac{2 + \sqrt{3}}{\sqrt{2^2 - \sqrt{3}^2}}  = \pi - arctg (2 + \sqrt{3}) = \frac{\pi}{2} + arcctg(2 + \sqrt{3}) }$ ,

т.к. $\large{arctg \ \alpha + arcctg \ \alpha = \frac{\pi}{2}, \ \forall \alpha}$.

Применим формулу котангенса половинного аргумента:

$\large{ctg \frac{\alpha}{2} = \frac{1 + cos \ \alpha}{sin \ \alpha} }$

$\large{arcctg(2 + \sqrt{3}) = arcctg \frac{1 + \frac{\sqrt{3}}{2}} {\frac{1}{2}} = arcctg \frac{1 + cos \frac{\pi}{6}}{sin \frac{\pi}{6}} = arcctg(ctg \frac{\pi}{12}) = \frac{\pi}{12} }$

$\large{\varphi_1 = \frac{\pi}{2} + \frac{\pi}{12} = \frac{7\pi}{12} }$


---

$Z_2 = \sqrt{3} + i$

$r_2 = |Z_2| = \sqrt{\sqrt{3}^2 + 1^2} = \sqrt{4} = 2$

$\sqrt{3} > 0; 1 > 0 \implies \varphi_2 \in I $ четверти.

$\large{\varphi_2 = arctg \frac{1}{\sqrt{3}} = \frac{\pi}{6}}$

---

$Z_3 = -1 + i$

$r_3 = |Z_3| = \sqrt{(-1)^2 + 1^2} = \sqrt{2}$

$-1 < 0; 1 > 0 \implies \varphi_3 \in II $ четверти.

$\large{\varphi_3 = \pi - arctg \frac{1}{1} = \pi - \frac{\pi}{4} = \frac{3\pi}{4}}$

### II.

Возведение комплексного числа в степень в тригонометрической форме:
    
$z^k = [r * (\cos \varphi + i \sin \varphi)]^n = r^n * [\cos(n\varphi) + i \sin(n\varphi) ]$

Умножение комплексных чисел в тригонометрической форме:
    
$z_1 * z_2 = r_1 * (\cos \varphi_1 + i \sin \varphi_1) *  r_2 * (\cos \varphi_2 + i \sin \varphi_2) = r_1r_2 * [\cos(\varphi_1 + \varphi_2) + i \sin(\varphi_1 + \varphi_2) ]$

Деление комплексных чисел в тригонометрической форме:

$\large{\frac{z_1}{z_2} = \frac{r_1 * (\cos \varphi_1 + i \sin \varphi_1)} { r_2 * (\cos \varphi_2 + i \sin \varphi_2)} = \frac {r_1}{r_2} }$$* [\cos(\varphi_1 - \varphi_2) + i \sin(\varphi_1 - \varphi_2) ]$

Вычислим $Z_1^K * Z_2^L, \ Z_2^M \ / \ Z_3^N$  в тригонометрической форме, используя приведенные формулы.

$\large{Z_1 ^ K = 2 ^ {10} * (\cos (10 * \frac{7\pi}{12}) + i \sin(10 * \frac{7\pi}{12})) = 2 ^ {10} * (\cos \frac{35\pi}{6} + i \sin \frac{35\pi}{6}) =}$

$\large{= 2 ^ {10} * (\cos (6\pi - \frac{\pi}{6}) + i \sin (6\pi - \frac{\pi}{6})) =  2 ^ {10} * (\cos \frac{\pi}{6} + i \sin(- \frac{\pi}{6}))}$

---

$\large{Z_2 ^ L = 2 ^ 6 * (\cos (6 * \frac{\pi}{6}) + i \sin(6 * \frac{\pi}{6})) = 2 ^ 6 * (\cos \pi + i \sin \pi) }$

---

$\large{Z_1 ^ K * Z_2 ^ L = 2 ^ {10} * 2 ^ 6 * (\cos (\pi + \frac{\pi}{6}) + i \sin(\pi - \frac{\pi}{6})) = 2 ^ {16} * (\cos (-\frac{\pi}{6}) + i \sin\frac{\pi}{6}) = }$

$\large{= - 2 ^ {15} \sqrt{3} + 2 ^ {15} i = -32768 \sqrt{3} + 32768 i }$

---

$\large{Z_2 ^ M = 2 ^ {15} * (\cos (15 * \frac{\pi}{6}) + i \sin(15 * \frac{\pi}{6})) = }$

$\large{= 2 ^ {15} * (\cos (2\pi + \frac{\pi}{2}) + i \sin (2\pi + \frac{\pi}{2})) =  2 ^ {15} * (\cos \frac{\pi}{2} + i \sin \frac{\pi}{2})}$

---

$\large{Z_3 ^ N = (\sqrt 2 )^ {27} * (\cos (27 * \frac{3\pi}{4}) + i \sin(27 * \frac{3\pi}{4})) = (\sqrt 2) ^ {27} * (\cos \frac{81\pi}{4} + i \sin \frac{81\pi}{4}) = }$

$\large{= (\sqrt 2) ^ {27} *  (\cos (20\pi + \frac{\pi}{4}) + i \sin (20\pi + \frac{\pi}{4})) = (\sqrt 2) ^ {27} *  (\cos \frac{\pi}{4} + i \sin \frac{\pi}{4}) }$

---

$\large{Z_2 ^ M / Z_3 ^ N = \frac{2 ^ {15}} {(\sqrt{2}) ^ {27}} * (\cos (\frac{\pi}{2} - \frac{\pi}{4}) + i \sin(\frac{\pi}{2} - \frac{\pi}{4})) = }$

$\large{=  \big(\sqrt{2}\big) ^ {30 - 27} * (\cos \frac{\pi}{4} + i \sin \frac{\pi}{4}) = \big(\sqrt{2}\big) ^ 3 * (\frac{\sqrt{2} } {2} + i \frac{\sqrt{2} } {2}) = 2 + 2 i}$
